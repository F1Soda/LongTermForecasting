# Дополнительные материалы


## Почему бы не делать Short-Term Forecasting?

Мы могли бы подумать, что сегодня мы стали куда более рациональными благодаря современной науке и технологиям. Тем не менее фондовые рынки кажутся более волатильными и иррациональными, чем когда-либо. Всё больше и больше акций торгуются продвинутыми компьютерными программами с краткосрочными стратегиями. Когда человек или машина пытаются предсказать краткосрочное движение рынка, они по сути пытаются угадать, что сделают другие участники — люди или алгоритмы. Это похоже на бешеную карусель, где каждый стремится опередить остальных. В конце концов это — бесполезное занятие!

Но конечно, есть некоторые модели, построенные на нейросетях. Из тех которые я слышал были основаны на RNN -- Реккурентные нейронные сети
Ещё конечно есть олдсукльные модели, по типу ARIMA, которые не обучаются, а являются подобием регрессионных моделей. Данные нужно брать от 3-х лет для лучшего предсказания
Есть вот такая статья на хабре https://habr.com/ru/articles/821231/, посвящённая ARIMА

## Почему P/Sales?

Быстрый ответ: технический это проще и для **нашей** модели предсказания будут лучше

Долгий ответ:

Если вы заметили, то для нашей формулы вообще говоря можно использовать разные мультпликаторы, из которых можно получить значение стоимости акции. То есть мы могли бы взять P/E:

$$
P/E\ (\text{Price\ to\ Earnings}) =  \frac{\text{Share\ Price}_t \cdot \text{Total\ Assets}_t}{\text{Earnings\ Per\ Share}_t \cdot \text{Total\ Assets}_t} = \frac{\text{Share\ Price}_t}{\text{Earnings\ Per\ Share}_t}.
$$

да мы впринципе могли бы использовать любой возможный мультипликатор. Например $P/T$ где $T$ — температура в селе Гафури. Но согласитесь, врядли этот мультипликатор несёт в себе какой-то глубоки смысл при анализе ценности компании.

Выбрали мы $\text{P/Sales}$ по той причине, что $\text{P/Earnings}$ может быть отрицательным(Прибыль отрицательная) и тогда не очень понятно как брать корень. Конечно, если вы знаете, что у компании каждый год только положительные доходы и в будущем таковые тоже ожидаются, то можно использовать и этот мультипликатор 


Если мы захотим использовать другие мультпликаторы, то нам либо придётся как дообрабатывать данные, либо использовать другие модели.

### Почему $\text{Ann Return}_t$ обозначает доходность через год $t + \text{Year}$, а не в момент времени $t$?

В действительности мы могли бы считать, что $\text{Ann Return}_t$ обозначает доходность, которую мы получим в момент времени $t$. И это было бы интуитивно понятнее.

Однако становится не понятно, как обозначить доход, который мы получим спустя 3-5 лет? $\text{Ann Return}_{t + 3, t + 4, t + 5}$? Да и как строить исторический график с такими параметрами?

Поэтому удобнее конкретному моменту времени $t$ сопостовлять будущий доход, который мы сможем получить для определённого промежутка времени. 

## А ещё?

А ещё мы не обсудили некоторые технические аспекты:
- Можно использовать более упрощённую формулу для подсчета среднего доходности. Для этого воспользуемся следующим фактом. При $X, Y, Z$ близких к $0$ справидливо 

    $$
    X + Y + Z \simeq (X + 1)(Y + 1)(Z + 1) - 1
    $$

    В нашей формуле все 3 параметра в общем случае достаточно малы, тогда мы можем записать

    $$
    E[\text{Ann\ Return}_t] \simeq E[\text{Dividend\ Yield}] + E[\text{Sales\ Pre\ Share\ Growth\ Rate}] + E\left[\left(\frac{\text{P/Sales}}{\text{P/Sales}_t}\right)^{1/\text{Years}}\right] - 1
    $$

    Эта формула конечно имеет больше ошибок, но для быстрой оценки пойдет. Её применение можно посмотреть в главе 3 Case Study: Tesla

- Доходы с акций и дивиденды облагаться налогами, а наша формула этого не учитывает. Мы не можем просто так добавить в нашу формулу зависимость от налогов по многим причинам: правило сбора налогов может меняться время от времени, сбор осуществляется в различные моменты времени. Где то налоги платятся по годам, а где то по кварталам. Даже если мы запишим формулу для какого то конкретного случая, то это не будет нести определённого смысла для использования. Подробные размышления можно почитать в 2.4 Tax

- Инфляция. Есть несколько подходов чтобы учитывать инфляцию:
    1) Делать поправки данных с учетом возможной инфляции
    2) Использовать $CPI$ — Индекс потребительских цен

    Сделать поправку не так и тяжело. Автор подробно рассматривает использование второго подхода в разделе 2.7 Inflation

- Обратный выкупи vs Дивиденды. Почему важно вкладывать дивиденды в покупку акций

    В современной экономике по мимо выплат дивидендов существует ещё много различных способов вознаграждения акционеров компании. Одна из самых популярных — обратный выкуп — когда компания выкупает свои акции. Выкуп обычно производят чтобы
    1. Увеличить стоимость акции
    2. Показать рынку, что акции недооценены
    3. Избежать налогов.

    Наша модель использует доходность дивидендов, которые у значительного количества акций отсутсвует. Как же это повлияет на нашу модель? Короткий ответ: никак, если мы используем правильно исторические данные.

    В нашей моделе все полученные денги от дивидендов сразу вкладываются в покупку новых акций. Поэтому соотношение владеимых нами акций увеличивается.

    А вот если компания не выплачивает дивиденды, то метрика $E[\text{Dividend Yield} + 1]$ становится равной 1. И вот тут можно задаться вопросом: окей, а как же наша модель будет учитывать обратные выкупы? Ответ кроется в параметре $Sales\ Per\ Share$: так как акций на рынке становится меньше, то прибыль приходимая на одну акцию возврастает.

    Именно поэтому разницы между дивидендами и обратными выкупами отсутсвует. Нужно лишь правильно обрабатывать исторические данные. Подробнее об этом можно почитать в 14.7 Share Buybacks vs. Dividends

### Случайное блуждание

Хорошо, значит полагаться на прошлое не получится. Но что если попытаться представить стоимость акций с помощью случайных блужданий?

Но в действительности акции скорее поддаются полу-случаынм блужданиям:
- Стоимости акций не могут в какой то момент стать больше чем ВВП страны. Так же сомнительно что она опустится до нулевых значений
- Стоимости акций часто реагируют на какие то изменения состояния в эконимике, которые мы можем отслеживать(Увидим в следующей главе)

Полагаясь на данные факты, попробуем описать модель предсказания стоимости цен, считая что они изменяются полу-случайным блужданиями

#### Формула для случайных блужданий 

Есть множество способов генерировать случайные блуждания для стоимости акций. Мы будем использовать следующий подход:

Пусть доходность акции, обозначаемые как $\text{RandomReturn}_t$, $IID$ — независимы и одинаково распределены.

Тогда стоимость акции $\text{RandomPrice}_{t+1}$ выражается как

$$
\text{RandomPrice}_{t+1} = \text{RandomPrice}_{t} \cdot \text{RandomReturn}_{t}.
$$

Будем полагать, что в момент времени $t$ доходность распрасределена нормально

$$
\text{RandomReturn}_{t} \sim N(\mu, \sigma^2).
$$

#### Формула для полу-случайных блужданий

Для полу-случайных блужданий мы будем использовать одну хитрость: Мы будем подмешивать в $\mu$ данные о настоящей стоимости акции.

То есть в каждый момент времени у нас будет новая матожа $\mu_t$, которую мы будем считать основываясь на некотором предположении о будущей стоимости акции:

$$
\mu_t = \left(\frac{\text{PredictionPrice}_{t + K}}{\text{RandomPrice}_{t}}\right)^{1/K},
$$

где $K$ — временной период, для которого мы делаем предположение.

#### Но откуда взять $\text{PredictionPrice}_{t + K}$?

На данный момент, мы пока не имеем какой-то теории как получить предсказание о будущей стоимости акции. Поэтому немного считерим: будем подставлять настоящее значение стоимости акции.

$$
\mu_t =  \left(\frac{\text{Price}_{t + K}}{\text{RandomPrice}_{t}}\right)^{1/K}
$$

#### Практика

Посмотрим, как себя будет вести стоимости акций. Сравним реальные данные с полностью случайным и полу-случайного блужданием. График будем строить для нормированной стоимости, то есть будем считать, что начальная стоимость равна $1$

По графику видно, что случайное и полу-случайное блуждание в целом угадала тренд акций Apple.

Синий график(случайные блуждания) имеет очень большой разнос по значеням: в каких то случаях получилось так, что доходность акций Apple превзошла общую доходность США, а в каких то уходила в минус.

Зелёный график(полу-случайные блуждания) ужее более точно описывает настоящее поведение акций. Однако и он не смог уловить локальные колебания в некоторых периодах длинной в год



## Почему угадывание значений мультипликаторов лучше, чем пытаться угадать стоимость акции?

Я сформулирую такую мысль: 

Мультипликаторы позволяют вынести какую то часть информации из большого датасета чисел(стомость акции, дивиденды, доходности, количество акций и прочие). Их суть в том, что они позволяют вывести какую то причину того, почему акция ведёт себя на рынке так, а не иначе. Но они не дают кокретного ответа  — перспективная ли акция или нет. Например может быть компания с значением $\text{P/Sales}$ скажем $0.5$, но это не значит что нужно срочно вкладывать все деньги в неё в надежде на будущий рост стоимости акции. Вполне возможно что компания собирается выжить максимум в последнии годы своей жизни, продовая все свои технологии и сокращая расходы. Может быть так, что компания не собирается закрываться, показывает низкий $\text{P/Sales}$ с выскомии доходами, однако эти доходы не идут на рост компании, в следствии чего возможна будущая стогнация с понижением стоимости акции. Или вот та же Тесла. Она показывала убытки многие годы, имела низкие показали, однако наращивала потенциал, что в итоге вылилось в огромные доходы.
 
Мультипликаторы должны лишь подкреплять наши предположения о ценности акции. Они не являются показателями ценности.



## Что, если перейти к прогнозированию по диапазону?

На графиках 5-ти и 8-ми лет иногда видно явные выбросы исторических данных для точек с низкими значениями P/Sales: фактическая доходность оказалась значительно выше, чем предсказывала модель. Такое обычно происходит в тех случаях, когда на начало определённого периода приходилось очень низкое значение P/Sales, а к концу — очень высокое (или наоборот). В результате для этого периода получается аномально высокая или низкая доходность, превращающаяся в выброс на точечной диаграмме. Что можно сделать - это перейти к рассмотрению прогноза на определённый период. Усреднение по всем возможным периодам внутри диапазона нивелирует такие выбросы, и разборс точек на графике становится меньше. Такое сглаживание достигается за счёт усреднения по диапазону инвестиционных горизонтов.

При этом формулу мы продолжаем использовать ту же самую:

$$
E[\text{Ann Return}_t] = \frac{a}{\text{P/Sales}_t^{1/\text{Years}}} - 1,
$$

Но в качестве $\text{Years}$ мы берём как среднее значение диапазона. Но при сравнении коэффициент $R^2$ заметно растёт: при рассмотрении трёхлетнего периода $R^2$=0.64, а на промежутке - уже 0.78.
